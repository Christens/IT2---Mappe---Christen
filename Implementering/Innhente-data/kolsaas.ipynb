```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)


#for i in range(1,len(tur)-1):
    #print(tur[i][1])
```

```python
import matplotlib.pyplot as plt
import numpy as np

hoyde = []
tid = []

for i in range(1,len(tur)-1):
    hoyde.append(float(tur[i][3]))
    tid.append(float(tur[i][0]))
    
plt.plot(tid,hoyde)
plt.show()
```

```python
lat = []
lon = []

for i in range(1,len(tur)-1):
    lat.append(float(tur[i][1]))
    lon.append(float(tur[i][2]))
    
plt.plot(lat,lon)
plt.show()
```
